```python
class top_5:
    def top_5amount_lifted(data):
        data1=data['Amount Lifted (kg)'].sort_values().tail().index
        data2=data.loc[data.index[data1]]
        print(data2[['Lifter Name','Amount Lifted (kg)']])
    data=pd.read_csv('powerlifting_dataset.csv')
    top_5amount_lifted(data)
```

```python
#age above 50 and box plot
class above_n:
    def fifty(data):
        maximumage=data['Age'].max()
        print('enter the age',end="")
        n=int(input())
        if n > maximumage:
            print("age above",maximumage,"data is not available")
        else:
            ageabove50=data[data['Age']>n]
            print(ageabove50)
            sns.boxplot(ageabove50)
            plt.title('data of age above 50')
            plt.xlabel('Total Lift Amount')
    data=pd.read_csv('powerlifting_dataset.csv')
    fifty(data)
    
```
