```python
df.sort_values('Date', inplace=True)

# Add new column for Last5_fulltimeHomeGoals
df['FTHG_L5'] = 0

# Calculate Last5_fulltimeHomeGoals
for i, row in df.iterrows():
    team = row['HomeTeam']
    date = row['Date']
    
    # Filter last 5 matches of the team before the current match
    last_matches = df[
        ((df['HomeTeam'] == team) | (df['AwayTeam'] == team)) & 
        (df['Date'] < date)
    ].tail(5)
    
    # Sum the goals scored by the team in those matches
    goals = 0
    for _, match in last_matches.iterrows():
        if match['HomeTeam'] == team:
            goals += match['FTHG']
        elif match['AwayTeam'] == team:
            goals += match['FTAG']
    
    df.at[i, 'FTHG_L5'] = goals

print(df)
```
